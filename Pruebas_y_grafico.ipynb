```python
from graphviz.backend import version
import numpy as np
import graphviz as gv
import json

class Nodo:
  def __init__(self,idx,lat,lon):
    l_idx =idx.replace('node/','')
    self.idx = l_idx
    self.lat = lat
    self.lon = lon
  def Return_idx(self):
    return self.idx
  def Return_Para_Print(self):
    return self.idx,self.lat,self.lon

Nodos = []

with open('parte_ohio.geojson') as f:
    data = json.load(f)


for feature in data['features']:
    Nodos.append(Nodo(feature['id'],feature['geometry']['coordinates'][0],feature['geometry']['coordinates'][1]))
  #print(feature['id'],feature['geometry']['coordinates'])

for Nodo in range(30):
  print(Nodos[Nodo].Return_idx(), Nodos[Nodo].lat, Nodos[Nodo].lon)
  #Solo est√° mostrando los 30 primeros 

graph = gv.Digraph("Prueba")
for u in range(len(Nodos)):
  graph.node(str(u),Nodos[u].Return_idx())

#print(graph)
#graph
```
